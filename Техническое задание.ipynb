**Задание №2**

```python
import numpy as np
```

```python
# Создаем Датафрейм из исходного файла
df2 = pd.read_excel('DataFrame.xlsx', sheet_name='2', header=5) 
df2 = df2.loc[0:9, 1:3]
df2
```

```python
# Заменяем отрицательные значения на NaN
def negative_to_nan(df2):
    for i in range(1, 4):
        df2.loc[(df2[i].values < 0), i] = np.nan
    
    return df2
```

```python
# Считаем среднее по каждому массиву
def mean_plus(df2):
    df = negative_to_nan(df2)
    
    print(df.mean())
```

```python
mean_plus(df2)
```

**Задание №3**

```python
# Создаем Датафрейм из исходного файла
df3 = pd.read_excel('DataFrame.xlsx', sheet_name='3', header=1, index_col='№ п/п')
df3 = df3.loc[:, 'Модель':'Сумма заказа']
df3.head(5)
```

```python
# Находим сумму заказа по каждой сделке
df3['Сумма заказа'] = df3['Цена за а/м']*df3['Кол-во автомобилей в заказе']
df3.head(5)
```

```python
# Вычисляем необходимые данные и записываем в новую таблицу
df3_sum = df3.groupby('Модель')[['Кол-во автомобилей в заказе','Сумма заказа']].agg('sum')
df3_sum
```

```python
# Переименовываем столбцы согласно исходному файлу
df3_sum.rename(columns = {'Кол-во автомобилей в заказе' : 'Количество проданных автомобилей',
                          'Сумма заказа' : 'Суммарные продажи по моделям'}, inplace = True)
df3_sum
```

```python
# Импортируем Датафрейм в таблицу xlsx
df3.to_excel('Задание 3.1.xlsx')
```

```python
# Импортируем Датафрейм в таблицу xlsx
df3_sum.to_excel('Задание 3.2.xlsx')
```

**Задание №4**

```python
# Создаем Датафрейм из первой таблицы исходного файла
df4 = pd.read_excel('DataFrame.xlsx', sheet_name='4', header=1, index_col='Модель')
df4 = df4.loc[:, '№':'Сумма заказа']
df4.head(5)
```

```python
# Создаем Датафрейм из второй таблицы исходного файла
df4_1 = pd.read_excel('DataFrame.xlsx', sheet_name='4', header=1)
df4_1 = df4_1.loc[0:12, ['Модель.1', 'Стоимость']]
df4_1
```

```python
# Переименовываем столбцы согласно первому датафрейму
df4_1.rename(columns = {'Модель.1' : 'Модель',
                          'Стоимость' : 'Цена за а/м'}, inplace = True)
df4_1
```

```python
# Присваиваем столбцу 'Модель' индекс
df4_1.set_index('Модель')
```

```python
# Объединяем датафреймы
new_df_4 = df4.merge(df4_1, on='Модель')
new_df_4.head(5)
```

```python
# Удаляем лишний столбец
del new_df_4['Цена за а/м_x']
```

```python
new_df_4.head(5)
```

```python
# Возвращаем исходное название столбцу 'Цена за а/м'
new_df_4.rename(columns = {'Цена за а/м_y' : 'Цена за а/м'}, inplace = True)
new_df_4.head(5)
```

```python
# Находим сумму заказа по каждой сделке
new_df_4['Сумма заказа'] = new_df_4['Цена за а/м']*new_df_4['Кол-во автомобилей в заказе']
new_df_4.head(5)
```

```python
# Применяем исходной порядок столбцов для датафрейма
new_df_4 = new_df_4[['№', 'Модель', 'Число', 'Месяц', 'Год', 'Квартал', 'Менеджер', 
                     'Заказчик', 'Цена за а/м', 'Кол-во автомобилей в заказе', 'Сумма заказа']]
new_df_4.head(5)
```

```python
# Сортируем по столбцу '№'
new_df_4 = new_df_4.sort_values(by='№')
new_df_4.head(5)
```

```python
# Присваиваем столбцу '№' индекс
new_df_4 = new_df_4.set_index('№')
new_df_4.head(5)
```

```python
# Создаем сводную таблицу по указанным условиям задачи
df4_pivot_table = pd.pivot_table(new_df_4, index=['Квартал', 'Заказчик'], values=['Сумма заказа'], aggfunc=[np.mean])
```

```python
# Переводим экспоненциальную запись в стандартный формат
df4_pivot_table = df4_pivot_table.style.format({'Сумма заказа': '{:.2f}'})
```

```python
df4_pivot_table
```

```python
# Импортируем Датафрейм в таблицу xlsx
new_df_4.to_excel('Задание 4.1.xlsx')
```

```python
# Импортируем Датафрейм в таблицу xlsx
df4_pivot_table.to_excel('Задание 4.2.xlsx')
```
